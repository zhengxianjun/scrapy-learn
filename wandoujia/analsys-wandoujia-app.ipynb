```python
#导入第三方包
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 读取app文件
app_info = pd.read_excel(r'app.xlsx')
```

```python
# 删除重复值
app_info.drop_duplicates(inplace=True)
app_info
```

```python
# 删除评论人数为-1的值
app_info_clean = app_info.loc[app_info.comments !=-1,] 
```

```python
app_info.describe(include=['object'])
```

```python
# 自定义函数，处理安装人数的单位
def to_install(x):
    if x.find('亿') !=-1:
        y = float(x[:-1])*10000
    elif x.find('万') != -1:
        y = float(x[:-1])
    else:
        y = float(x[:-1])/10000
    return(y)
```

```python
# 安装人数转换
app_info_clean['install_new'] = app_info_clean.install.apply(to_install)
app_info_clean.head()
```

```python
# 自定义匿名函数
y = lambda x: float(x[:-2]) if x.find('MB') !=-1 else float(x[:-2])/1024
```

```python
app_info_clean['size_new'] = app_info_clean['size'].apply(y)
app_info_clean
```

```python
# 自定义匿名函数，将暂无设置为缺失值
y1 = lambda x: np.nan if x == '暂无' else float(x[:-1])/100
app_info_clean['love_new'] = app_info_clean['love'].apply(y1)
app_info_clean
```

```python
# 用中位数对好评率进行填补
app_info_clean['love_new'] = app_info_clean.love_new.fillna(app_info_clean['love_new'].median())
app_info_clean
```

```python
app_info_clean['update_new'] = pd.to_datetime(app_info_clean['update'],format='%Y年%m月%d日')
app_info_clean
```
